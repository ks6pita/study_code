```python
#
# 必要なモジュールの読み込み


import pandas as pd
import numpy as np

import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.vector_ar.var_model import VAR
from statsmodels.tsa.vector_ar.svar_model import SVAR

from graphviz import Digraph

import warnings
warnings.simplefilter('ignore')

import matplotlib.pyplot as plt
plt.style.use('ggplot') #グラフのスタイル
plt.rcParams['figure.figsize'] = [8, 6] # グラフサイズ設定
plt.rcParams['font.size'] = 10 #フォントサイズ
```

```python
#
# 必要なデータセット（時系列データ）の読み込み
#

dataset='MMM.csv' #データセットのファイル名
df = pd.read_csv(
    dataset,
    parse_dates=True,
    index_col='day')

# print(df)#確認
df
```

```python
df.index
```

```python
df.plot()
plt.show()
```

```python
for i in range(len(df.columns)):
    
    #Augmented Dickey-Fuller 単位根検定
    c_results = adfuller(df.iloc[:,i],regression='c')  
    
    #p値出力
    print(df.columns[i],'\t',c_results[1]) 
```

```python
#
# VARモデルの構築
#

# 最大のラグ数
maxlags = 10

# モデルのインスタンス生成
var_model = VAR(df)

# 最適なラグの探索
lag = var_model.select_order(maxlags).selected_orders
print('最適なラグ：',lag['aic'],'\n')

# モデルの学習
results = var_model.fit(lag['aic'])

# 結果出力
print(results.summary())
```

```python
#
# グレンジャー因果性の検定
#   帰無仮説：グレンジャー因果なし
#   対立仮説：グレンジャー因果あり
#

for i in range(len(df.columns)):
    for j in range(len(df.columns)):
        if i != j :
            #因果の検定
            test_results = results.test_causality(causing=i, caused=j) 
            # #p値
            # test_results.pvalue
            #検定結果の出力
            print(df.columns[i],'->',df.columns[j],'\t',test_results.pvalue)
```

```python
from graphviz import Digraph

# インスタンスの生成
graph = Digraph()

# グラフにノードを追加
for col in df.columns:
    graph.node(col)

# 因果を検定し有意なとき線を引く
for i in df.columns:
    for j in df.columns:
        if i != j:
            # 因果の検定
            test_results = results.test_causality(causing=i, caused=j)
            # p値が0.05未満のとき線を追加
            if test_results.pvalue < 0.05:
                graph.edge(i, j)
            
# 有向グラフを表示
graph
```

```python
#
# インパルス応答関数(直交化)
#

irf = results.irf()
irf.plot(orth=True)
plt.show()
```

```python
#
# SVARモデルの構築
#

#　目的変数の行列Aの定義
A_0 = np.array([[1, 'E','E'],
                ['E', 1,'E'], 
                ['E','E', 1]])

# モデルのインスタンス生成
SVAR_model = SVAR(df, svar_type='A', A=A_0)

# 学習
results = SVAR_model.fit(maxlags=3)

#　推定結果（目的変数の行列A）
results.A
```

```python
#
# インパルス応答関数(構造)
#

irf = results.irf()
irf.plot()
plt.show()
```
